## <u>Visualizaciones y Análisis </u>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Subir datos
df = pd.read_csv('/amazon_products_sales_data.csv')
df.head()
```

```python
df.info()
```

## <u> Valores Nulos </u>

```python
# Asumiendo que 'missing_value' es la serie que obtuviste
missing_value = df.isna().sum().sort_values(ascending=False)

# Eliminar las columnas que no tienen valores faltantes para una mejor visualización
missing_value = missing_value[missing_value > 0]

# Crear el gráfico de barras
plt.figure(figsize=(12, 7))
ax = sns.barplot(x=missing_value.index, y=missing_value.values, palette='viridis')

# Agregar etiquetas a las barras
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center',
                va = 'center',
                xytext = (0, 9),
                textcoords = 'offset points')

# Personalizar el gráfico
plt.title('Valores Faltantes por Columna', fontsize=16)
plt.xlabel('Columnas', fontsize=12)
plt.ylabel('Cantidad de Valores Faltantes', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Análisis Univariado
###  Distribución de Valoraciones
(Busco entender la satisfaccion general de los clientes con los productos)

```python
# 1. Contar la frecuencia de cada valoración
rating_counts = df['product_rating'].value_counts().sort_index()

# 2. Crear el gráfico de barras
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=rating_counts.index, y=rating_counts.values, palette='viridis')

# 3. Personalizar el gráfico
plt.title('Distribución de Valoraciones de Productos', fontsize=16)
plt.xlabel('Puntuación de Valoración', fontsize=12)
plt.ylabel('Número de Productos', fontsize=12)
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)

# 4. Agregar etiquetas numéricas a las barras para mayor precisión
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10,
                xytext=(0, 5), textcoords='offset points')

plt.tight_layout()
plt.show()

```

```python
### Distribución de Productos por Categoría
```

```python
# 1. Contar el número de productos en cada categoría
category_counts = df['product_category'].value_counts()

# 2. Seleccionar las top 10 categorías para mejorar la visualización
top_10_categories = category_counts.head(10)

# 3. Crear el gráfico de barras
plt.figure(figsize=(14, 8))
ax = sns.barplot(x=top_10_categories.values, y=top_10_categories.index, palette='Spectral')

# 4. Personalizar el gráfico
plt.title('Top 10 Categorías de Productos por Cantidad', fontsize=16)
plt.xlabel('Cantidad de Productos', fontsize=12)
plt.ylabel('Categoría del Producto', fontsize=12)

# 5. Agregar etiquetas numéricas a las barras
for p in ax.patches:
    width = p.get_width()
    plt.text(width + 10,  # Posición X (un poco a la derecha de la barra)
             p.get_y() + p.get_height() / 2, # Posición Y (en el centro vertical de la barra)
             f'{int(width)}', # Texto a mostrar (el valor de la barra)
             ha='left',
             va='center')

plt.tight_layout()
plt.show()
```

## Análisis Bivariado
### Impacto del Precio con Descuento
(Busco entender la eficienia de los descuentos)

```python
# Asumiendo que las columnas de tu DataFrame se llaman 'original_price',
# 'discounted_price' y 'bought_in_last_month'

# 1. Crear una nueva columna para identificar si hay descuento o no
df['has_discount'] = (df['original_price'] != df['discounted_price']) & (~df['discounted_price'].isna())

# 2. Reemplazar los valores booleanos por etiquetas más descriptivas
df['has_discount'] = df['has_discount'].map({True: 'Con Descuento', False: 'Sin Descuento'})

# 3. Crear el diagrama de cajas
plt.figure(figsize=(10, 6))
sns.boxplot(x='has_discount', y='purchased_last_month', data=df, palette='pastel')

# 4. Personalizar el gráfico
plt.title('Impacto del Descuento en las Ventas', fontsize=16)
plt.xlabel('Tipo de Producto', fontsize=12)
plt.ylabel('Número de Unidades Vendidas (Último Mes)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

##  Descuento vs. Ventas
(impacto de los descuentos)

```python
# 1. Calcular el porcentaje de descuento
# Asegurarse de evitar divisiones por cero o valores nulos en el precio original
df['porcentaje_descuento'] = 100 * (1 - df['discounted_price'] / df['original_price'])

# 2. Crear el diagrama de dispersión
plt.figure(figsize=(12, 7))
sns.scatterplot(
    data=df,
    x='porcentaje_descuento',
    y='purchased_last_month',
    alpha=0.6,
    s=15 # Tamaño de los puntos
)

# 3. Personalizar el gráfico
plt.title('Relación entre el Porcentaje de Descuento y las Ventas', fontsize=16)
plt.xlabel('Porcentaje de Descuento (%)', fontsize=12)
plt.ylabel('Número de Unidades Vendidas (Último Mes)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

## Análisis Multivariado

### Precio Promedio, Valoración Promedio y Cantidad de Productos


```python
# 1. Agrupar los datos por categoría y calcular las métricas promedio
# Reemplaza 'categoria_del_producto' con el nombre de tu columna de categorías
# Reemplaza 'precio_actual' y 'valoracion_del_producto' con los nombres de tus columnas
category_metrics = df.groupby('product_category').agg(
    precio_promedio=('discounted_price', 'mean'),
    valoracion_promedio=('product_rating', 'mean'),
    cantidad_productos=('product_category', 'count')
).reset_index()

# 2. Filtrar para mostrar solo las categorías con más productos (ejemplo: top 10)
top_categories = category_metrics.sort_values(by='cantidad_productos', ascending=False).head(10)

# 3. Crear el diagrama de burbujas
plt.figure(figsize=(14, 8))
sns.scatterplot(
    data=top_categories,
    x='precio_promedio',
    y='valoracion_promedio',
    size='cantidad_productos',  # El tamaño de la burbuja representa la cantidad de productos
    hue='product_category', # El color representa la categoría
    palette='Spectral',
    sizes=(100, 1000),  # Rango del tamaño de las burbujas
    alpha=0.8
)

# 4. Personalizar el gráfico
plt.title('Relación entre Precio, Valoración y Cantidad por Categoría', fontsize=16)
plt.xlabel('Precio Promedio ($)', fontsize=12)
plt.ylabel('Valoración Promedio', fontsize=12)
plt.legend(title='Categoría', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.show()
```

## Outliers Detection

```python
original_size = df.shape[0]
print(f"Tamaño original del dataset: {original_size}")
```

```python
column_to_check = 'original_price'

# 1. Calcular el primer cuartil (Q1) y el tercer cuartil (Q3)
Q1 = df[column_to_check].quantile(0.25)
Q3 = df[column_to_check].quantile(0.75)

# 2. Calcular el rango intercuartílico (IQR)
IQR = Q3 - Q1

# 3. Definir los límites para detectar outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 4. Encontrar los outliers
outliers = df[(df[column_to_check] < lower_bound) | (df[column_to_check] > upper_bound)]
iqr_size = outliers.shape[0]
iqr_removed = original_size - iqr_size

# 5. Imprimir la cantidad de outliers encontrados
print(f"Número de outliers en la columna '{column_to_check}': {len(outliers)}")
print(f"Tamaño final: {iqr_removed}")
print(f"Registros eliminados: {iqr_size} ({100 * iqr_size/original_size:.2f}%)")
```
